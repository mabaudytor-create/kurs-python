```python
import math
r=input("Wpisz promień koła")
r=float(r)
Pole=math.pi*r**2
print(f'Pole: {Pole}')
obwód=2*math.pi*r
print(f'Obwód: {obwód}')
```

```python
x1=input("Wpisz współrzędną x1")
x1=float(x1)
y1=input("Wpisz współrzędną y1")
y1=float(y1)
x2=input("Wpisz współrzędną x2")
x2=float(x2)
y2=input("Wpisz współrzędną y2")
y2=float(y2)
odległość=math.sqrt((x2-x1)**2+(y2-y1)**2)
print(f'Odległość: {odległość}')
```
